```python
import pandas as pd
import psycopg2

conn = psycopg2.connect(database='Cars_DB',user='postgres', password='postgres', host='localhost', port='5432')
cursor = conn.cursor()


sql = '''CREATE TABLE employee(
 id  SERIAL NOT NULL,
 name varchar(20) not null,
 state varchar(20) not null
)'''
cursor.execute(sql)
data = [('Babita', 'Bihar'), ('Anushka', 'Hyderabad'), 
        ('Anamika', 'Banglore'), ('Sanaya', 'Pune'),
        ('Radha', 'Chandigarh')]

for d in data:
    cursor.execute("INSERT into employee(name, state) VALUES (%s, %s)", d)
    print("List has been inserted to employee table successfully...")
conn.commit()
 
conn.close()
```

```python
import pandas as pd
import psycopg2
import csv

conn = psycopg2.connect(database='Cars_DB',user='postgres', password='postgres', host='localhost', port='5432')
cursor = conn.cursor()
with open('./source/Stackline_inhouse_share.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader) # Skip the header row.
    records=0
    for row in reader:
        records+1
        cursor.execute("insert into stack_line_inhouse_share (RETAILER_ID,RETAILER_NAME,RETAILER_SKU,WEEK_ID) values (%s,%s,%s,%s)",row)
    
    conn.commit()
    
    conn.close() 
    print('\n{} Records are transfered' .format(records))
```
